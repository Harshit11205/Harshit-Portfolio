```python
df = df.dropna()
df.shape
```

```python
df.head(10)
```

```python
#converting date time format
df["date_added"] = pd.to_datetime(df['date_added'],format="mixed")
df['day_added'] = df['date_added'].dt.day
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month
df['year_added']=df['year_added'].astype(int);
df['day_added']=df['day_added'].astype(int);
```

```python
df.head(10)
```

```python
#visualizing the type
import seaborn as sns
import matplotlib.pyplot as plt

# Create a countplot for the 'type' column
sns.set(style="whitegrid")
plt.figure(figsize=(5, 5))  # Set the figure size

sns.countplot(x='type', data=df)  # Create the countplot

plt.title('Type')
plt.show()  # Show the plot
```

```python
#visualizing ratings
import seaborn as sns
import matplotlib.pyplot as plt

# Create a countplot for the 'rating' column
sns.set(style="whitegrid")
plt.figure(figsize=(8, 5))  # Set the figure size

# Create the countplot and rotate x-axis labels
ax = sns.countplot(x='rating', data=df, order=df['rating'].value_counts().index)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="right")

plt.title('Rating')
plt.show()  # Show the plot
```

```python
#visualizing the relation between type and rating
plt.figure(figsize=(10,5))
sns.countplot(x='rating',hue='type',data=df)
plt.title('Realtion between Type and Rating')
plt.show()
```

```python
#visualizing types using pie chart
labels=['Movie','TV show']
size= df['type'].value_counts()
colors= plt.cm.Wistia(np.linspace(0,1,2))
explode= [0,0.1]
plt.rcParams['figure.figsize'] = (10,6)
plt.pie(size,labels=labels, colors = colors, explode = explode, shadow = True, startangle = 90)
plt.title('Distribution of Type',fontsize=25)
plt.legend()
plt.show()
```

```python
#PIE-CHART FOR RATING
df['rating'].value_counts().plot.pie(autopct='%1.1f%%', shadow=True,figsize=(20,10))
plt.show()
```
